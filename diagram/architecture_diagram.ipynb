```python
from diagrams import Diagram, Cluster, Edge
from diagrams.aws.storage import SimpleStorageServiceS3
from diagrams import Cluster, Diagram
from diagrams.programming.language import Python
from urllib.request import urlretrieve
from diagrams.custom import Custom
from diagrams.digitalocean.storage import Folder
from diagrams.oci.connectivity import DNS
from diagrams.elastic.beats import Filebeat
from diagrams.generic.database import SQL
```

```python
graph_attr = {
  "width": "1000",
  "height": "1000"
}

with Diagram("\n\nArchitecture Diagram \n\n\nWorkflow : Clean and Validate Data \nTransform Data in DBT", show=False, graph_attr=graph_attr, direction="LR") as diag:
  with Cluster("\nRAW Data"):
    local_storage = Folder("\nStorage")

  with Cluster("\nLocal Storage"):
    local_storage_2 = Folder("\nTemporary\nStorage")

  with Cluster("\Data Cleaning"):
    py_cleaner = Python("\nData Cleaner")
    py_validator = Python("\nData validator")

  py_dataloader = Python("\nData Loader")
  py_dataloader_2 = Python("\nData Loader")
  s3_bucket = SimpleStorageServiceS3("\nS3 Bucket")

  snowflake_db_url = "https://pbs.twimg.com/media/EDT-rsuWwAAmhav.png"
  snowflake_db_icon = "elastic.png"
  urlretrieve(snowflake_db_url, snowflake_db_icon)
  snowflake_db = Custom("Snowflake\nData Warehouse", snowflake_db_icon, edge_label=" ", x_pos="1", y_pos="1")

  with Cluster("\DBT transformation"):
    sql_1 = SQL("\n Transform Data")
    sql_2 = SQL("\n Deploy")

  local_storage >> py_cleaner
  py_cleaner >> py_validator
  py_validator >> py_cleaner

  py_cleaner >> local_storage_2 >> py_dataloader >> s3_bucket

  local_storage_2 >> py_dataloader_2 >> snowflake_db


  snowflake_db >> sql_1 >> sql_2 >> snowflake_db




diag
```
